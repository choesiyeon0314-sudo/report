```python
#ì§‘í•© A = {1,2,3,4,5} ì— ê´€í•œ ê´€ê³„í–‰ë ¬ì„ ì…ë ¥ë°›ì•„ì„œ, ì´ ê´€ê³„ê°€ ë™ì¹˜ ê´€ê³„ì„ì„ íŒë³„í•˜ê³ ,
#ë™ì¹˜ ê´€ê³„ì´ë©´ ë™ì¹˜ë¥˜ë¥¼ ì¶œë ¥í•˜ì‹œì˜¤.
A = [1, 2, 3, 4, 5]
domain = len(A)


#í–‰ë ¬ ì¶œë ¥
def matrixout(mx):
    print("\nâ”Œ" + " " * (domain * 2 + 1) + "â”")
    for i in range(1, domain + 1):
        print("| ", end="")
        for j in range(1, domain + 1):
            print(f"{mx[i][j]} ", end="")
        print("|")
    print("â””" + " " * (domain * 2 + 1) + "â”˜")


#ê´€ê³„ ìˆœì„œìŒ ì¶œë ¥
def print_relation(mx):
    print("\n[ê´€ê³„ Rì— í¬í•¨ëœ ìˆœì„œìŒ]")
    for i in range(1, domain + 1):
        for j in range(1, domain + 1):
            if mx[i][j] == 1:
                print(f"({A[i-1]}, {A[j-1]})", end=' ')
    print("\n")


# ì¸ì ‘ ë¦¬ìŠ¤íŠ¸ ì¶œë ¥
def graphout(mx):
    print("\n[ê´€ê³„ì˜ ê·¸ë˜í”„ í‘œí˜„ (ì¸ì ‘ ë¦¬ìŠ¤íŠ¸)]")
    for i in range(1, domain + 1):
        print(f"{A[i-1]} -> ", end='')
        lst = [A[j-1] for j in range(1, domain + 1) if mx[i][j] == 1]
        if lst:
            print(", ".join(str(x) for x in lst))
        else:
            print("âˆ…")
    print()


#êµ¬ì„±ì¡°ê±´ 1. ê´€ê³„í–‰ë ¬ ì…ë ¥ ê¸°ëŠ¥
def input_relation():
    mx = [[0] * (domain + 1) for _ in range(domain + 1)]
    print("ì§‘í•© A =", A)
    print(f"{domain} x {domain} ì •ë°©í–‰ë ¬ì„ í–‰ ë‹¨ìœ„ë¡œ ì…ë ¥í•˜ì‹œì˜¤.\n")

    for i in range(1, domain + 1):
        while True:
           #ì‚¬ìš©ìë¡œë¶€í„° ì •ë°©í–‰ë ¬ì„ í–‰ ë‹¨ìœ„ë¡œ ì…ë ¥ë°›ìŒ
            row = input(f"{i}ë²ˆì§¸ í–‰ ì…ë ¥ (0/1 ë‹¤ì„¯ ê°œ): ").split()
            row = list(map(int, row))
            for j in range(1, domain + 1):
              #2ì°¨ì›ë°°ì—´ë¡œ ì €ì¥
                mx[i][j] = row[j - 1]
            break
    return mx


# êµ¬ì„± ì¡°ê±´ 2. ë™ì¹˜ê´€ê³„íŒë³„ ê¸°ëŠ¥ .  (ë°˜ì‚¬ / ëŒ€ì¹­ / ì¶”ì´) -> ë™ì¹˜ê´€ê³„
def is_reflexive(mx):
    for i in range(1, domain + 1):
        if mx[i][i] == 0:
            return False
    return True


def is_symmetric(mx):
    for i in range(1, domain + 1):
        for j in range(1, domain + 1):
            if mx[i][j] == 1 and mx[j][i] == 0:
                return False
    return True


def is_transitive(mx):
    for i in range(1, domain + 1):
        for j in range(1, domain + 1):
            if mx[i][j] == 1:
                for k in range(1, domain + 1):
                    if mx[j][k] == 1 and mx[i][k] == 0:
                        return False
    return True


def is_equivalence(mx):
    return is_reflexive(mx) and is_symmetric(mx) and is_transitive(mx)


# êµ¬ì„±ì¡°ê±´ 3. ë™ì¹˜ê´€ê³„ì¼ ê²½ìš° ë™ì¹˜ë¥˜ ì¶œë ¥ ê¸°ëŠ¥
def equivalence_classes(mx):
    print("\n[ë™ì¹˜ë¥˜]")
    for a in range(1, domain + 1):
        print(f"[{A[a-1]}] = {{ ", end='')
        elements = []
        for b in range(1, domain + 1):
            if mx[a][b] == 1:
                elements.append(str(A[b-1]))
        print(", ".join(elements), "}")
    print()


#êµ¬ì„±ì¡°ê±´ 4. íí¬ êµ¬í˜„ 0 ê¸°ëŠ¥. íí¬ (ë°˜ì‚¬ / ëŒ€ì¹­ / ì¶”ì´)

def reflexive_closure(mx):
    r = [row[:] for row in mx]
    for i in range(1, domain + 1):
        r[i][i] = 1
    return r


def symmetric_closure(mx):
    r = [row[:] for row in mx]
    for i in range(1, domain + 1):
        for j in range(1, domain + 1):
            if r[i][j] == 1 or r[j][i] == 1:
                r[i][j] = r[j][i] = 1
    return r


def transitive_closure(mx):
    r = [row[:] for row in mx]
    for k in range(1, domain + 1):
        for i in range(1, domain + 1):
            for j in range(1, domain + 1):
                if r[i][k] == 1 and r[k][j] == 1:
                    r[i][j] = 1
    return r


# íí¬ ì¶œë ¥ + íí¬ í›„ ë™ì¹˜ê´€ê³„ íŒë³„ + ë™ì¹˜ë¥˜ ì¶œë ¥
def show_closure(title, original, closure_func):
    print("\n==============================")
    print(f"[{title}]")

    print("\në³€í™˜ ì „ ê´€ê³„í–‰ë ¬:")
    matrixout(original)

    transformed = closure_func(original)

    print("\në³€í™˜ í›„ ê´€ê³„í–‰ë ¬:")
    matrixout(transformed)

    print("\në³€í™˜ í›„ ê·¸ë˜í”„ í‘œí˜„:")
    graphout(transformed)

    print("\níí¬ í›„ ë™ì¹˜ ê´€ê³„ íŒë³„:", end=" ")
    if is_equivalence(transformed):
        print("ì˜ˆ (ë™ì¹˜ ê´€ê³„)")
        equivalence_classes(transformed)
    else:
        print("ì•„ë‹ˆì˜¤ (ë™ì¹˜ ê´€ê³„ ì•„ë‹˜)")

    return transformed


#-------------------------------------
def main():
    mx = input_relation()

    print("\n[ì…ë ¥ëœ ê´€ê³„í–‰ë ¬]")
    matrixout(mx)
    print_relation(mx)
    graphout(mx)

    print("[ê´€ê³„ì˜ ì„±ì§ˆ íŒë³„]")
    print("ë°˜ì‚¬ ê´€ê³„?  :", "ì˜ˆ" if is_reflexive(mx) else "ì•„ë‹ˆì˜¤")
    print("ëŒ€ì¹­ ê´€ê³„?  :", "ì˜ˆ" if is_symmetric(mx) else "ì•„ë‹ˆì˜¤")
    print("ì¶”ì´ ê´€ê³„?  :", "ì˜ˆ" if is_transitive(mx) else "ì•„ë‹ˆì˜¤")

    # ë™ì¹˜ê´€ê³„
    if is_equivalence(mx):
        print("\nâ‡’ ì´ ê´€ê³„ëŠ” ë™ì¹˜ ê´€ê³„ì…ë‹ˆë‹¤.")
        equivalence_classes(mx)
    else:
        print("\nâ‡’ ì´ ê´€ê³„ëŠ” ë™ì¹˜ ê´€ê³„ê°€ ì•„ë‹™ë‹ˆë‹¤.")

    # ğŸ”¥ íí¬ (ë°˜ì‚¬ / ëŒ€ì¹­ / ì¶”ì´) â€” ë³€í™˜ ì „/í›„ + íí¬ í›„ ë™ì¹˜ë¥˜ ì¶œë ¥
    if not is_reflexive(mx):
        show_closure("ë°˜ì‚¬ íí¬", mx, reflexive_closure)

    if not is_symmetric(mx):
        show_closure("ëŒ€ì¹­ íí¬", mx, symmetric_closure)

    if not is_transitive(mx):
        show_closure("ì¶”ì´ íí¬", mx, transitive_closure)


if __name__ == "__main__":
    main()
```
